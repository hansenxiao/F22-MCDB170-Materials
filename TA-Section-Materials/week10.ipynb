## solve_ivp
```
scipy.integrate.solve_ivp(fun, t_span, y0)
```

This function numerically integrates a system of ordinary differential equations given an initial value:


```
dy / dt = f(t, y)
y(t0) = y0
```

## Example
1. Basic exponential functions showing automatically chosen time points.
```
dy[0]/dt = -0.5*y[0]
dy[1]/dt = 0.5*y[1]
```

````python
import numpy as np
from scipy.integrate import solve_ivp

def exponential_decay(t, y): 
    return np.array([-0.5 * y[0], 0.5*y[1]])

sol = solve_ivp(exponential_decay, [0, 10], [10, 0.1])
print(sol.t)
print(sol.y)

import matplotlib.pyplot as plt
plt.plot(sol.t, sol.y[0], 'r')
plt.plot(sol.t, sol.y[1])
plt.xlabel('t')
plt.legend(['y0 = 2', 'y1 = 4'], shadow=True)
plt.show()
````



2. Specifying tspan.
```
tspan = np.linspace(0,10,1000)
```

````python
sol = solve_ivp(exponential_decay, [0, 10], [10, 0.1],
                t_eval=np.linspace(0,10,1000))
````
