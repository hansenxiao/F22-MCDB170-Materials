## solve_ivp
```
scipy.integrate.solve_ivp(fun, t_span, y0)
```

This function numerically integrates a system of ordinary differential equations given an initial value:


```
dy / dt = f(t, y)
y(t0) = y0
```

## Example
1. Basic exponential decay showing automatically chosen time points.

````python
from scipy.integrate import solve_ivp
def exponential_decay(t, y): return -0.5 * y
sol = solve_ivp(exponential_decay, [0, 10], [2, 4, 8])
print(sol.t)
print(sol.y)
````



2. Specifying points where the solution is desired.

````python
sol = solve_ivp(exponential_decay, [0, 10], [2, 4, 8],
                t_eval=[0, 1, 2, 4, 10])
print(sol.t)
print(sol.y)
````



3. matplotlib

````python
import matplotlib.pyplot as plt
plt.plot(sol.t, sol.y[0], 'r')
plt.plot(sol.t, sol.y[1])
plt.plot(sol.t, sol.y[2])
plt.xlabel('t')
plt.legend(['y0 = 2', 'y0 = 4', 'y0 = 8'], shadow=True)
plt.show()
````
